# At first glance we achieved what we needed but variable merged_researchers has a lot of NaN values which endangers our future work. So, here comes in handy data imputation methods. And the most appropriate here is:
# 1) To convey clustering analysis for each year;
# 2) To find out relation between counries and clusters;
# 3) To find mean value for each cluster;
# 4) Fill in missing values 

```python
from sklearn.cluster import KMeans
from sklearn import preprocessing
standart_2015 = preprocessing.scale(year_2015_thebest.set_index('Country Name'))
standart_2016 = preprocessing.scale(year_2016_thebest.set_index('Country Name'))
standart_2017 = preprocessing.scale(year_2017_thebest.set_index('Country Name'))
standart_2018 = preprocessing.scale(year_2018_thebest.set_index('Country Name'))
```

```python
year_2015_standart = pd.DataFrame(standart_2015)
year_2016_standart = pd.DataFrame(standart_2016)
year_2017_standart = pd.DataFrame(standart_2017)
year_2018_standart = pd.DataFrame(standart_2018)
```

```python
K = range(1,11)
models = [KMeans(n_clusters=k, random_state=42).fit(year_2015_standart) for k in K]
dist = [model.inertia_ for model in models]
plt.plot(K, dist, marker='o')
plt.title('The elbow showing the optimal number of k')
plt.xlabel('k')
plt.ylabel('Distance')
plt.show()
```

```python
model = KMeans(n_clusters=3, random_state=42)
model.fit(year_2015_standart)
year_2015_thebest['cluster'] = model.labels_
print(year_2015_thebest)
```

```python
K = range(1,11)
models = [KMeans(n_clusters=k, random_state=42).fit(year_2016_standart) for k in K]
dist = [model.inertia_ for model in models]
plt.plot(K, dist, marker='o')
plt.title('The elbow showing the optimal number of k')
plt.xlabel('k')
plt.ylabel('Distance')
plt.show()
```

```python
model = KMeans(n_clusters=3, random_state=42)
model.fit(year_2016_standart)
year_2016_thebest['cluster'] = model.labels_
print(year_2016_thebest.groupby('cluster').mean())
```

```python
K = range(1,11)
models = [KMeans(n_clusters=k, random_state=42).fit(year_2017_standart) for k in K]
dist = [model.inertia_ for model in models]
plt.plot(K, dist, marker='o')
plt.title('The elbow showing the optimal number of k')
plt.xlabel('k')
plt.ylabel('Distance')
plt.show()
```

```python
model = KMeans(n_clusters=3, random_state=42)
model.fit(year_2017_standart)
year_2017_thebest['cluster'] = model.labels_
print(year_2017_thebest.groupby('cluster').mean())
```

```python
K = range(1,11)
models = [KMeans(n_clusters=k, random_state=42).fit(year_2018_standart) for k in K]
dist = [model.inertia_ for model in models]
plt.plot(K, dist, marker='o')
plt.title('The elbow showing the optimal number of k')
plt.xlabel('k')
plt.ylabel('Distance')
plt.show()
```

```python
model = KMeans(n_clusters=3, random_state=42)
model.fit(year_2016_standart)
year_2016_thebest['cluster'] = model.labels_
print(year_2016_thebest.groupby('cluster').mean())
```
